### 3.1 Handling Missing Data ü§î

*   **Goal:**
    
    *   Identify missing values and decide whether to impute (fill in) or remove them.
        
*   **Steps & Code:**

```python
# Check for missing values in each column
missing_counts = df.isnull().sum()
print("Missing values in each column:\n", missing_counts[missing_counts > 0])
```

**Imputation Example:**

For numerical features, you might use the median, and for categorical features, the mode.

```python
# For numerical columns: Impute with median
num_cols = df.select_dtypes(include=['number']).columns
for col in num_cols:
    df[col] = df[col].fillna(df[col].median())

# For categorical columns: Impute with mode
cat_cols = df.select_dtypes(include=['object']).columns
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

**Visualization:**

Plot a bar chart of missing values before and after imputation (if desired).

```python
import matplotlib.pyplot as plt

# Plot missing values before imputation (if you had stored it)
missing_before = missing_counts[missing_counts > 0]
missing_before.plot(kind='bar', title='Missing Values Before Imputation', figsize=(8, 5))
plt.ylabel('Count')
plt.show()
```

```python
# Count missing values after imputation
missing_after = df.isnull().sum()

# Check if there are any missing values left after imputation
if missing_after[missing_after > 0].empty:
    print("No missing values left after imputation.")
else:
    # Plot missing values after imputation if any
    plt.figure(figsize=(8, 5))
    missing_after[missing_after > 0].plot(kind='bar', color='green', alpha=0.7)
    plt.title('Missing Values After Imputation')
    plt.ylabel('Count')
    plt.show()
```

*   **Goal:**

    *   Interaction terms: For instance, combining GrLivArea and OverallQual (if these features exist) might capture a compound effect.

### 3.4 Encoding Categorical Variables üìè

* Since regression models require numeric inputs, we convert all categorical variables to dummy variables.

```python
df_encoded = pd.get_dummies(df, drop_first=True)
print("Data after encoding:")
print(df_encoded.head())
print(df_encoded.info())
```

### 3.5 Data Scaling üìè

*   **Goal:**
    
    *   Scale numerical features to a similar range‚Äîespecially important for algorithms like SVR and KNN.
        


```python
from sklearn.preprocessing import StandardScaler

# Select numerical columns for scaling (excluding the target if already transformed)
num_features = df_encoded.select_dtypes(include=['number']).columns
scaler = StandardScaler()

# Scale the features; create a new DataFrame to hold scaled values
df_scaled = df_encoded.copy()
df_scaled[num_features] = scaler.fit_transform(df[num_features])

# Visualize scaled features (e.g., histogram of a scaled feature)
plt.figure(figsize=(8, 5))
sns.histplot(df_scaled['Gr Liv Area'], kde=True, color='purple')
plt.title('Distribution of Scaled GrLivArea')
plt.xlabel('Scaled GrLivArea')
plt.ylabel('Frequency')
plt.show()
```

### 3.5 Train-Test Split üö¶

*   **Goal:**
    
    *   Split the dataset into training and testing sets to evaluate model performance on unseen data.

```python
from sklearn.model_selection import train_test_split

# Define feature set X and target variable y
# Here we assume 'SalePrice_log' as the target; change to 'SalePrice' if preferred.
X = df_scaled.drop(['SalePrice', 'SalePrice_log'], axis=1, errors='ignore')
y = df_scaled['SalePrice_log'] if 'SalePrice_log' in df_scaled.columns else df_scaled['SalePrice']

# Split data: 80% training, 20% testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

# Print shapes to confirm the split
print("Training set shape:", X_train.shape)
print("Test set shape:", X_test.shape)
```

4\. Model Building and Regularization üèóÔ∏è
-----------------------------------------

### 4.1 Baseline Model: Simple Linear Regression

1.  **Train a basic Linear Regression model** using scikit-learn‚Äôs LinearRegression.
    
2.  **Evaluate** the model using metrics such as R¬≤ score or MSE on the test set.
    
3.  **Compare** results with regularized models later.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Baseline Model: Linear Regression
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)

# Predictions on test set
y_pred_lr = lr_model.predict(X_test)

# Evaluate
mse_lr = mean_squared_error(y_test, y_pred_lr)
r2_lr = r2_score(y_test, y_pred_lr)

print("üîπ Linear Regression Results:")
print(f"   MSE: {mse_lr:.2f}")
print(f"   R¬≤ : {r2_lr:.2f}")
```

### 4.2 Regularized Models üîí

Regularization helps **prevent overfitting** by penalizing large coefficients. We‚Äôll explore three common types:

1.  **Ridge (L2)**
    
2.  **Lasso (L1)**
    
3.  **ElasticNet (L1 + L2)** (optional)
    

#### 4.2.1 Ridge Regression (L2)

*   **Key Parameter:** alpha
    
    *   Higher alpha ‚Üí Stronger regularization (coefficients shrink more).

```python
from sklearn.linear_model import Ridge

# Ridge Regression
ridge_model = Ridge(alpha=10)  # Example alpha=10
ridge_model.fit(X_train, y_train)

# Predictions
y_pred_ridge = ridge_model.predict(X_test)

# Evaluation
mse_ridge = mean_squared_error(y_test, y_pred_ridge)
r2_ridge = r2_score(y_test, y_pred_ridge)

print("\nüîπ Ridge Regression Results:")
print(f"   Alpha: {ridge_model.alpha}")
print(f"   MSE: {mse_ridge:.2f}")
print(f"   R¬≤ : {r2_ridge:.2f}")
```

#### 4.2.2 Lasso Regression (L1)

*   **Key Parameter:** alpha
    
    *   Higher alpha ‚Üí More coefficients become zero (feature selection effect).

```python
from sklearn.linear_model import Lasso

# Lasso Regression
lasso_model = Lasso(alpha=0.01)  # Example alpha=0.01
lasso_model.fit(X_train, y_train)

# Predictions
y_pred_lasso = lasso_model.predict(X_test)

# Evaluation
mse_lasso = mean_squared_error(y_test, y_pred_lasso)
r2_lasso = r2_score(y_test, y_pred_lasso)

print("\nüîπ Lasso Regression Results:")
print(f"   Alpha: {lasso_model.alpha}")
print(f"   MSE: {mse_lasso:.2f}")
print(f"   R¬≤ : {r2_lasso:.2f}")
```

#### 4.2.3 (Optional) ElasticNet (L1 + L2)

*   **Key Parameters:** alpha and l1\_ratio
    
    *   l1\_ratio = 0 ‚Üí Pure Ridge, = 1 ‚Üí Pure Lasso, values in between ‚Üí Mix.

```python
from sklearn.linear_model import ElasticNet

# ElasticNet
elastic_model = ElasticNet(alpha=0.1, l1_ratio=0.5)  # Example alpha, l1_ratio
elastic_model.fit(X_train, y_train)

# Predictions
y_pred_elastic = elastic_model.predict(X_test)

# Evaluation
mse_elastic = mean_squared_error(y_test, y_pred_elastic)
r2_elastic = r2_score(y_test, y_pred_elastic)

print("\nüîπ ElasticNet Results:")
print(f"   Alpha: {elastic_model.alpha}")
print(f"   L1 Ratio: {elastic_model.l1_ratio}")
print(f"   MSE: {mse_elastic:.2f}")
print(f"   R¬≤ : {r2_elastic:.2f}")
```

### 4.3 Visualizing Model Coefficients üìä

Comparing model coefficients can help you understand how each feature influences the predictions‚Äîespecially for linear-type models.

1.  **Extract** coefficients from each model.
    
2.  **Plot** them side by side or individually.

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_top_coefficients(coefs, feature_names, model_name, top_n=10):
    """
    Plot top_n features (by absolute coefficient) for a given model.
    """
    # Sort by absolute value of coefficients
    sorted_idx = np.argsort(np.abs(coefs))[-top_n:]
    top_features = feature_names[sorted_idx]
    top_coefs = coefs[sorted_idx]

    plt.figure(figsize=(6, 4))
    plt.barh(top_features, top_coefs, color='skyblue', alpha=0.7)
    plt.title(f"{model_name} - Top {top_n} Coefficients (by |value|)")
    plt.axvline(x=0, color='black', linewidth=0.8)
    plt.xlabel("Coefficient Value")
    plt.tight_layout()
    plt.show()

# Example usage:
plot_top_coefficients(lr_model.coef_, X_train.columns, "LinearRegression", top_n=10)
plot_top_coefficients(ridge_model.coef_, X_train.columns, "Ridge", top_n=10)
plot_top_coefficients(lasso_model.coef_, X_train.columns, "Lasso", top_n=10)
plot_top_coefficients(elastic_model.coef_, X_train.columns, "ElasticNet", top_n=10)
```

5\. Advanced Model Training with Hyperparameter Tuning ‚öôÔ∏è
---------------------------------------------------------

In this section, we construct pipelines that seamlessly combine preprocessing (scaling) with modeling, then use hyperparameter optimization (via RandomizedSearchCV or GridSearchCV) along with k‚Äëfold cross-validation (e.g., k=5) to robustly tune our models. We also tune regularization parameters (e.g., alpha in Ridge and Lasso) as needed.

> **Note:** Make sure you have already preprocessed your data in Section 4 so that all features are numeric and scaled, and then split into X\_train, X\_test, y\_train, y\_test.

### 5.1 Pipeline Setup & Hyperparameter Tuning for Each Model

#### **(A) Ridge Regression (L2 Regularization)**

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import RandomizedSearchCV
from sklearn.linear_model import Ridge

# Pipeline: scaling -> Ridge Regression
ridge_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('ridge', Ridge())
])

# Define parameter distribution for Ridge (random search)
param_distributions_ridge = {
    'ridge__alpha': [0.0001, 0.001, 0.01, 0.1, 1, 10, 50, 100, 200, 500]
}

# Use RandomizedSearchCV instead of GridSearchCV
ridge_search = RandomizedSearchCV(
    estimator=ridge_pipeline,
    param_distributions=param_distributions_ridge,
    n_iter=5,  # Only test 5 random combinations (can be increased)
    cv=5,
    scoring='neg_mean_squared_error',
    random_state=42,  # Ensures reproducibility
    n_jobs=-1
)

ridge_search.fit(X_train, y_train)

print("üîπ Ridge Regression Best Parameters:", ridge_search.best_params_)
print("üîπ Ridge Regression Best CV MSE:", -ridge_search.best_score_)
```

#### **(B) Lasso Regression (L1 Regularization)**

```python
from sklearn.linear_model import Lasso

# Create a pipeline: scaling -> Lasso Regression
lasso_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('lasso', Lasso(max_iter=10000))
])

# Define parameter grid for Lasso (tuning alpha)
param_distributions_lasso = {
    'lasso__alpha': [0.0001, 0.001, 0.01, 0.1, 1, 10, 50, 100]
}

lasso_search = RandomizedSearchCV(
    estimator=lasso_pipeline,
    param_distributions=param_distributions_lasso,
    n_iter=5,
    cv=5,
    scoring='neg_mean_squared_error',
    random_state=42,
    n_jobs=-1
)
lasso_search.fit(X_train, y_train)

print("üîπ Lasso Regression Best Parameters:", lasso_search.best_params_)
print("üîπ Lasso Regression Best CV MSE:", -lasso_search.best_score_)
```

#### **(C) Polynomial Regression**

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Define a list of selected features (adjust based on your EDA)
# Make sure these column names exactly match the names in your DataFrame
selected_features = ['GrLivArea_OverallQual', 'Overall Qual', 'Bldg Type_Twnhs', 'Neighborhood_NridgHt', '2nd Flr SF', 'Functional_Sal', 'Exterior 1st_BrkFace', 'Misc Feature_Gar2', 'MS Zoning_C (all)']  # Updated feature names

# Create a pipeline that selects only these features, scales them, applies polynomial expansion, then fits Linear Regression
poly_pipeline = Pipeline([
    ('selector', ColumnTransformer([
        ('select', 'passthrough', selected_features)
    ])),  # Added remainder='drop' to handle missing features
    ('poly', PolynomialFeatures(degree=2, include_bias=False)),
    ('linreg', LinearRegression())
])

# Define parameter grid (only tuning degree if desired)
param_distributions_poly = {
    'poly__degree': [2, 3]  # Try degree 2 and 3 on the reduced feature set
}

from sklearn.model_selection import RandomizedSearchCV

poly_search = RandomizedSearchCV(
    estimator=poly_pipeline,
    param_distributions=param_distributions_poly,
    n_iter=2,  # fewer iterations since the grid is small
    cv=5,
    scoring='neg_mean_squared_error',
    random_state=42,
    n_jobs=1  # use one job to save memory
)
poly_search.fit(X_train, y_train)
print("üîπ Polynomial Regression Best Parameters:", poly_search.best_params_)
print("üîπ Polynomial Regression Best CV MSE:", -poly_search.best_score_)
```

#### **(D) Support Vector Regression (SVR)**

```python
from sklearn.svm import SVR
from sklearn.model_selection import RandomizedSearchCV

# Pipeline: scaling -> SVR
svr_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('svr', SVR())
])

# Define parameter grid/distributions for SVR
param_distributions_svr = {
    'svr__C': [0.1, 1, 10, 100],
    'svr__epsilon': [0.01, 0.1, 0.5, 1],
    'svr__gamma': ['scale', 'auto'],
    'svr__kernel': ['linear', 'rbf', 'poly']
}

svr_search = RandomizedSearchCV(
    estimator=svr_pipeline,
    param_distributions=param_distributions_svr,
    n_iter=10,
    cv=5,
    scoring='neg_mean_squared_error',
    random_state=42,
    n_jobs=-1
)
svr_search.fit(X_train, y_train)

print("üîπ SVR Best Parameters:", svr_search.best_params_)
print("üîπ SVR Best CV MSE:", -svr_search.best_score_)
```

#### **(E) Decision Tree Regression**

```python
from sklearn.tree import DecisionTreeRegressor

# Pipeline: scaling -> Decision Tree Regressor
dt_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('dt', DecisionTreeRegressor())
])

# Define parameter grid for Decision Tree
param_grid_dt = {
    'dt__max_depth': [None, 5, 10, 20, 50],
    'dt__min_samples_split': [2, 5, 10],
    'dt__min_samples_leaf': [1, 2, 4],
    'dt__max_features': [None, 'auto', 'sqrt', 'log2']
}

dt_search = RandomizedSearchCV(
    estimator=dt_pipeline,
    param_distributions=param_grid_dt,
    n_iter=20,
    cv=5,
    scoring='neg_mean_squared_error',
    random_state=42,
    n_jobs=-1
)
dt_search.fit(X_train, y_train)

print("üîπ Decision Tree Best Parameters:", dt_search.best_params_)
print("üîπ Decision Tree Best CV MSE:", -dt_search.best_score_)
```

#### **(F) Random Forest Regression**

```python
from sklearn.ensemble import RandomForestRegressor

# Pipeline: scaling -> Random Forest Regressor
rf_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('rf', RandomForestRegressor())
])

# Define parameter grid for Random Forest
param_grid_rf = {
    'rf__n_estimators': [50, 100, 200, 300],
    'rf__max_depth': [None, 10, 20, 30],
    'rf__min_samples_split': [2, 5, 10],
    'rf__min_samples_leaf': [1, 2, 4],
    'rf__max_features': [None, 'auto', 'sqrt', 'log2']
}

rf_search = RandomizedSearchCV(
    estimator=rf_pipeline,
    param_distributions=param_grid_rf,
    n_iter=20,
    cv=5,
    scoring='neg_mean_squared_error',
    random_state=42,
    n_jobs=-1
)
rf_search.fit(X_train, y_train)

print("üîπ Random Forest Best Parameters:", rf_search.best_params_)
print("üîπ Random Forest Best CV MSE:", -rf_search.best_score_)
```

#### **(G) K‚ÄëNearest Neighbors Regression (KNN)**

```python
from sklearn.neighbors import KNeighborsRegressor

# Pipeline: scaling -> KNN Regressor
knn_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('knn', KNeighborsRegressor())
])

# Define parameter grid for KNN
param_grid_knn = {
    'knn__n_neighbors': [3, 5, 7, 9, 11],
    'knn__weights': ['uniform', 'distance'],
    'knn__p': [1, 2]  # 1: Manhattan, 2: Euclidean
}

knn_search = RandomizedSearchCV(
    estimator=knn_pipeline,
    param_distributions=param_grid_knn,
    n_iter=10,
    cv=5,
    scoring='neg_mean_squared_error',
    random_state=42,
    n_jobs=-1
)
knn_search.fit(X_train, y_train)

print("üîπ KNN Best Parameters:", knn_search.best_params_)
print("üîπ KNN Best CV MSE:", -knn_search.best_score_)
```

### 5.2 Summary and Explanation

*   **Pipelines:**For each regression model, we build a pipeline that includes a data scaling step (StandardScaler) followed by the regression estimator. This ensures that all preprocessing is applied consistently.
    
*   **Hyperparameter Optimization:**We define a parameter grid (or distributions) for each model‚Äôs key hyperparameters, such as:
    
    *   **Ridge & Lasso:** Tuning the regularization strength (alpha).
        
    *   **Polynomial Regression:** Tuning the degree of polynomial features.
        
    *   **SVR:** Tuning kernel type, C, epsilon, and gamma.
        
    *   **Decision Tree & Random Forest:** Tuning tree depth, minimum samples for split/leaf, and number of estimators (for RF).
        
    *   **KNN:** Tuning number of neighbors, distance weights, and distance metric (p).
        
*   **K‚ÄëFold Cross-Validation:** The cv parameter (set to 5 in these examples) ensures that each hyperparameter combination is evaluated using 5‚Äëfold cross‚Äëvalidation for robust performance estimates.
    
*   **Scoring:** We use neg\_mean\_squared\_error as our scoring metric (remember to convert it to positive MSE for interpretation).
    
*   **Search Strategy:** We employ GridSearchCV for exhaustive search when the parameter grid is small (e.g., Ridge, Lasso, Polynomial Regression) and RandomizedSearchCV for broader parameter spaces (e.g., SVR, Decision Tree, Random Forest, KNN).

### 6.1 Evaluation Metrics

We compute the following metrics for each model:

*   **Mean Squared Error (MSE)**
    
*   **Root Mean Squared Error (RMSE)**
    
*   **Mean Absolute Error (MAE)**
    
*   **R¬≤ Score**

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

def evaluate_model(model, X_test, y_test):
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    return mse, rmse, mae, r2

# Example: Using best estimators from hyperparameter tuning for each model:
# Assume these variables are defined from previous tuning steps:
#   ridge_search.best_estimator_, lasso_search.best_estimator_, poly_search.best_estimator_,
#   svr_search.best_estimator_, dt_search.best_estimator_, rf_search.best_estimator_, knn_search.best_estimator_

models = {
    "Ridge": ridge_search.best_estimator_,
    "Lasso": lasso_search.best_estimator_,
    "Polynomial": poly_search.best_estimator_,
    "SVR": svr_search.best_estimator_,
    "DecisionTree": dt_search.best_estimator_,
    "RandomForest": rf_search.best_estimator_,
    "KNN": knn_search.best_estimator_
}

# Evaluate each model and collect metrics
results = {}
for name, model in models.items():
    mse, rmse, mae, r2 = evaluate_model(model, X_test, y_test)
    results[name] = {
        "MSE": f"{mse:.2f}",
        "RMSE": f"{rmse:.2f}",
        "MAE": f"{mae:.2f}",
        "R¬≤": f"{r2:.2f}"
        }

# Create a summary DataFrame for easier comparison
import pandas as pd
results_df = pd.DataFrame(results).T
print("Evaluation Metrics for Each Model:")
print(results_df)
```
